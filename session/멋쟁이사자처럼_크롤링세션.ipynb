```python
from selenium import webdriver
from bs4 import BeautifulSoup

import time    #  time.sleep(대기시간)
import pandas as pd
```

```python
#Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.
 
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
#해당 url로 이동
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1" 
driver.get(url)
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

4. 원하는 정보 찾기




* soup.select('태그명')
* soup.select('.class속성값')
* soup.select('부모태그정보 > 자식태그정보')
* soup.select('#id속성값')



```python
channel_list = soup.select('tr')
len(channel_list)
```

```python
channel_list = soup.select('.aos-init')
len(channel_list)
```

```python
channel = channel_list[0]
channel
```

```python
# 채널명 찾기
title = channel.select('h1 > a')
len(title)
```

```python
title = channel.select('h1 > a')[0].text.strip()
title
```

```python
# 카테고리 정보 정리하기
view = channel.select('td.view_cnt')[0].text.strip()
view
```

```python
# 리스트에 100개 채널 정보 저장하기

results = []

for channel in channel_list:
    title = channel.select('h1 > a')[0].text.strip()
    category = channel.select('p.category')[0].text.strip()
    subscriber = channel.select('td.subscriber_cnt')[0].text.strip()
    view = channel.select('td.view_cnt')[0].text.strip()

    data = [title, category, subscriber, view]
    results.append(data)
results
```

```python
# pandas를 활용하여 엑셀 파일에 저장하기

# 표 모양을 DataFrame으로 바꾸기
df = pd.DataFrame(results)
df
```

```python
df.colums = ['title', 'category', 'subscriber', 'view']
df
```

```python
# DataFrame을 엑셀로 저장
df.to_excel('./youtube_rank_top100.xlsx', index=False)
```
